```python
import pandas as pd
import random

# 매핑 정의: 원본 업태명 → 가능한 분류 목록
category_mapping = {
    "한식": ["한식"],
    "호프/통닭": ["치킨", "족발/보쌈", "한식"],
    "기타": ["간식", "샐러드", "샌드위치", "디저트"],
    "식육(숯불구이)": ["구이", "한식"],
    "분식": ["분식", "1인분"],
    "중국식": ["중식"],
    "경양식": ["돈까스", "양식"],
    "통닭(치킨)": ["치킨"],
    "횟집": ["회/해물"],
    "일식": ["일식"],
    "정종/대포집/소주방": ["한식"],
    "탕류(보신용)": ["찜/탕", "한식"],
    "까페": ["커피/차"],
    "김밥(도시락)": ["도시락", "포장", "1인분"],
    "패스트푸드": ["버거", "포장", "1인분"],
    "뷔페식": ["피자", "양식", "멕시칸"],
    "외국음식전문점(인도,태국등)": ["아시안"],
    "냉면집": ["한식", "죽"],
    "복어취급": ["회/해물"],
    "패밀리레스트랑": ["양식", "멕시칸"],
    "감성주점": ["한식"],
    "전통찻집": ["커피/차", "디저트"]
}

# 매핑 함수 정의
def map_category(upname):
    candidates = category_mapping.get(upname)
    if candidates:
        return random.choice(candidates)
    else:
        print(f"Unknown category for 업태명: {upname}")
        return "기타"

# 적용
df = df.dropna(subset=["위생업태명"])
df["category"] = df["위생업태명"].apply(map_category)
```

```python
import pandas as pd
import random
import string

# username 생성 함수 (user0001, user0002 ...)
def generate_username(index):
    return f"user{index:04d}"

# 비밀번호 생성 함수 (영문+숫자+특수문자 포함, 10자 기본)
def generate_password(length=10):
    while True:
        password = ''.join(random.choices(
            string.ascii_letters + string.digits + "@$!%*?&", k=length
        ))
        # 조건 만족 여부 확인
        if (any(c.isalpha() for c in password) and
            any(c.isdigit() for c in password) and
            any(c in "@$!%*?&" for c in password)):
            return password

df["username"] = [generate_username(i) for i in range(len(df))]
df["password"] = [generate_password() for _ in range(len(df))]
```

## 로그인 및 회원가입

```python
import requests
import json

def login(row):
    login_url = "http://localhost:8080/api/login"

    request_data = {
        "username": row["username"],
        "password": row["password"]
    }

    try:
        response = requests.post(login_url, json=request_data)
        return response.headers.get('Access-Token', None)
    except Exception as e:
        print(f"[ERROR] {row['username']} 로그인 실패: {e}")
        return None


def signup_and_login(row):
    signup_url = "http://localhost:8080/api/owner/members/sign-up"

    token = login(row)
    if token is not None:
        return token # 이미 로그인된 경우 토큰 반환
    
    request_data = {
        "username": row["username"],
        "password": row["password"],
        "name": row["사업장명"],
        "nickname": row["사업장명"],
        "email": f"{row['username']}@example.com",
        "phone": "010-9999-9999"
    }

    try:
        response = requests.post(signup_url, json=request_data)
        return login(row)
    except Exception as e:
        print(f"[ERROR] {row['username']} 회원가입 실패: {e}")
        return None
```

## 가맹점 생성 요청 전송

```python
import requests
import json

def post_store(row):
    token = signup_and_login(row)
    if (token is None):
        print(f"[ERROR] {row['username']} 회원가입 또는 로그인 실패")
        return None

    url = "http://localhost:8080/api/owner/store-regist"

    # 랜덤한 배달비 설정
    random_fee = random.choice(range(1000, 3001, 500))

    store_request_data = {
        "name": row["사업장명"],
        "description": row["사업장명"],
        "address": row["도로명전체주소"],
        "lng": row["longitude"],
        "lat": row["latitude"],
        "phone": row["소재지전화"],
        "isFranchise": False,
        "categoryName": row["category"],
        "defaultDeliveryFee": random_fee,
        "onlyOneDeliveryFee": random_fee + 2000
    }

    files = {
        'request': ('request.json', json.dumps(store_request_data), 'application/json'),
        'images': ('store_image.txt', get_image_url(row["category"]), 'text/plain')  # 이미지 대신 문자열 파일
    }

    try:
        response = requests.post(url, files=files, headers={
            'Authorization': f'Bearer {token}',
        })

        if response.status_code == 201 or response.ok:
            response_data = response.json()
            store_id = response_data.get("data", {}).get("storeId")
            print(f"[SUCCESS] 매장 등록됨 - storeId: {store_id}")
            return store_id
        else:
            print(f"[ERROR] 매장 등록 실패 - 상태코드: {response.status_code}")
            return None

    except Exception as e:
        print(f"[ERROR] 매장 '{row['사업장명']}' 등록 실패: {e}")
        return None
```

## 메뉴 설정 요청 전송

### 메뉴 그룹 설정

```python
# 메뉴 그룹 생성 요청 JSON
categories = [
    "한식", "중식", "피자", "찜/탕", "치킨", "분식", "돈까스", "족발/보쌈",
    "구이", "일식", "회/해물", "양식", "커피/차", "디저트", "간식", "아시안",
    "샌드위치", "샐러드", "버거", "멕시칸", "도시락", "죽", "포장", "1인분"
]

payload_list = [
    {
        "menuGroups": [
            {
                "menuGroupName": category,
                "menuGroupPriority": 1,
                "menuGroupIsActive": True
            },
            {
                "menuGroupName": '사이드 메뉴',
                "menuGroupPriority": 2,
                "menuGroupIsActive": True
            }
        ]
    }
    for category in categories
]

def get_payload_by_category(category_name):
    try:
        index = categories.index(category_name)
        return payload_list[index]
    except ValueError:
        print(f"[ERROR] '{category_name}'는 유효한 카테고리가 아닙니다.")
        return None
```

```python
import requests
import json

def post_menu_group_by_category(row, token, store_id):
    url = f"http://localhost:8080/api/owner/{store_id}/menu-groups"

    category = row.get("category")

    payload = get_payload_by_category(category)
    if payload is None:
        return None

    headers = {
        "Content-Type": "application/json",
        "Authorization": f"Bearer {token}"
    }

    try:
        response = requests.post(url, json=payload, headers=headers)
        return response.status_code
    except Exception as e:
        print(f"[ERROR] {category} 등록 실패: {e}")
        return None
```

## 메뉴 추가

```python
import random
# 메뉴 추가 요청 JSON
categories = [
    "한식", "중식", "피자", "찜/탕", "치킨", "분식", "돈까스", "족발/보쌈",
    "구이", "일식", "회/해물", "양식", "커피/차", "디저트", "간식", "아시안",
    "샌드위치", "샐러드", "버거", "멕시칸", "도시락", "죽", "포장", "1인분"
]

def generate_option_group(name, required, priority):
    return {
        "optionGroupName": name,
        "isRequired": required,
        "minSelect": 1 if required else 0,
        "maxSelect": 1,
        "priority": priority,
        "options": [
            {
                "optionName": f"{name} A",
                "optionPrice": 0,
                "optionStatus": random.choices(
                    ["ON_SALE", "HIDDEN", "OUT_OF_STOCK"],
                    weights=[0.9, 0.05, 0.05],
                    k=1
                )[0],
                "optionPriority": 1
            },
            {
                "optionName": f"{name} B",
                "optionPrice": 500,
                "optionStatus": random.choices(
                    ["ON_SALE", "HIDDEN", "OUT_OF_STOCK"],
                    weights=[0.9, 0.05, 0.05],
                    k=1
                )[0],
                "optionPriority": 2
            }
        ]
    }

# 각 카테고리별 메뉴 payload 구성
menu_payload_list = []
for i, category in enumerate(categories):
    price = random.choice(range(6000, 24001, 500))
    payload = {
        "menuName": f"{category} 메뉴",
        "menuDescription": f"{category}의 대표 메뉴입니다.",
        "menuPrice": price,
        "menuStatus": "ON_SALE",
        "menuGroupName": category,
        "menuPriority": 1,
        "optionGroups": [
            generate_option_group("사이즈", required=True, priority=1),
            generate_option_group("추가 선택", required=False, priority=2)
        ]
    }
    menu_payload_list.append(payload)

# 카테고리명으로 payload 찾는 함수
def get_menu_payload_by_category(category_name):
    try:
        index = categories.index(category_name)
        return menu_payload_list[index]
    except ValueError:
        print(f"[ERROR] '{category_name}'는 유효한 카테고리가 아닙니다.")
        return None
```

```python
import requests
import json

def post_menu_by_category(row, token, store_id):
    url = f"http://localhost:8080/api/owner/{store_id}/menus/new"

    category = row.get("category")
    payload = get_menu_payload_by_category(category)
    if payload is None:
        return None

    headers = {
        "Authorization": f"Bearer {token}"
    }

    files = {
        'request': ('request.json', json.dumps(payload), 'application/json'),
        'images': ('menu_image.txt', get_image_url(row["category"]), 'text/plain')  # 실제 이미지 파일이면 바이너리로 열기
    }

    try:
        response = requests.post(url, headers=headers, files=files)
        return response.status_code
    except Exception as e:
        print(f"[ERROR] {category} 등록 실패: {e}")
        return None
    

def post_side_menu_by_category(row, token, store_id):
    url = f"http://localhost:8080/api/owner/{store_id}/menus/new"

    payload = {
        "menuName": f"사이드 메뉴",
        "menuDescription": f"사이드 메뉴입니다.",
        "menuPrice": price,
        "menuStatus": "ON_SALE",
        "menuGroupName": "사이드 메뉴",
        "menuPriority": 1,
        "optionGroups": []
    }

    headers = {
        "Authorization": f"Bearer {token}"
    }

    files = {
        'request': ('request.json', json.dumps(payload), 'application/json'),
        'images': ('menu_image.txt', get_image_url(row["category"]), 'text/plain')  # 실제 이미지 파일이면 바이너리로 열기
    }

    try:
        response = requests.post(url, headers=headers, files=files)
        return response.status_code
    except Exception as e:
        print(f"[ERROR] {category} 등록 실패: {e}")
        return None
```
