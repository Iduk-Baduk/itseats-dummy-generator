```python
import pandas as pd
import random

# 매핑 정의: 원본 업태명 → 가능한 분류 목록
category_mapping = {
    "한식": ["한식"],
    "호프/통닭": ["치킨", "족발/보쌈", "한식"],
    "기타": ["간식", "샐러드", "샌드위치", "디저트"],
    "식육(숯불구이)": ["구이", "한식"],
    "분식": ["분식", "1인분"],
    "중국식": ["중식"],
    "경양식": ["돈까스", "양식"],
    "통닭(치킨)": ["치킨"],
    "횟집": ["회/해물"],
    "일식": ["일식"],
    "정종/대포집/소주방": ["한식"],
    "탕류(보신용)": ["찜/탕", "한식"],
    "까페": ["커피/차"],
    "김밥(도시락)": ["도시락", "포장", "1인분"],
    "패스트푸드": ["버거", "포장", "1인분"],
    "뷔페식": ["피자", "양식", "멕시칸"],
    "외국음식전문점(인도,태국등)": ["아시안"],
    "냉면집": ["한식", "죽"],
    "복어취급": ["회/해물"],
    "패밀리레스트랑": ["양식", "멕시칸"],
    "감성주점": ["한식"],
    "전통찻집": ["커피/차", "디저트"]
}

# 매핑 함수 정의
def map_category(upname):
    candidates = category_mapping.get(upname)
    if candidates:
        return random.choice(candidates)
    else:
        print(f"Unknown category for 업태명: {upname}")
        return "기타"

# 적용
df = df.dropna(subset=["위생업태명"])
df["category"] = df["위생업태명"].apply(map_category)
```

```python
import pandas as pd
import random
import string

# username 생성 함수 (user0001, user0002 ...)
def generate_username(index):
    return f"user{index:04d}"

# 비밀번호 생성 함수 (영문+숫자+특수문자 포함, 10자 기본)
def generate_password(length=10):
    while True:
        password = ''.join(random.choices(
            string.ascii_letters + string.digits + "@$!%*?&", k=length
        ))
        # 조건 만족 여부 확인
        if (any(c.isalpha() for c in password) and
            any(c.isdigit() for c in password) and
            any(c in "@$!%*?&" for c in password)):
            return password

df["username"] = [generate_username(i) for i in range(len(df))]
df["password"] = [generate_password() for _ in range(len(df))]
```

```python
import requests
import json

def login(row):
    login_url = "http://localhost:8080/api/login"

    request_data = {
        "username": row["username"],
        "password": row["password"]
    }

    try:
        response = requests.post(login_url, json=request_data)
        return response.headers.get('Access-Token', None)
    except Exception as e:
        print(f"[ERROR] {row['username']} 로그인 실패: {e}")
        return None


def signup_and_login(row):
    signup_url = "http://localhost:8080/api/members/sign-up"

    token = login(row)
    if token is not None:
        return token # 이미 로그인된 경우 토큰 반환
    
    request_data = {
        "username": row["username"],
        "password": row["password"],
        "name": row["사업장명"],
        "nickname": row["사업장명"],
        "email": f"{row['username']}@example.com",
        "phone": "010-9999-9999"
    }

    try:
        response = requests.post(signup_url, json=request_data)
        return login(row)
    except Exception as e:
        print(f"[ERROR] {row['username']} 회원가입 실패: {e}")
        return None
```

```python
import random

# 카테고리 → 사진 디렉토리 매핑
category_to_dir = {
    "한식": "korean", "중식": "chinese", "피자": "pizza", "찜/탕": "korean", "치킨": "chicken",
    "분식": "korean", "돈까스": "japanese", "족발/보쌈": "korean", "구이": "korean", "일식": "japanese",
    "회/해물": "japanese", "양식": "western", "커피/차": "cafe", "디저트": "dessert", "간식": "dessert",
    "아시안": "asian", "샌드위치": "western", "샐러드": "salad", "버거": "american", "멕시칸": "mexican",
    "도시락": "mealbox", "죽": "korean", "포장": "mealbox", "1인분": "mealbox"
}

# 디렉토리별 최대 이미지 수
dir_max_image_count = {
    "korean": 139, "chinese": 16, "pizza": 76, "salad": 36, "dessert": 102, "japanese": 55,
    "western": 22, "american": 76, "mexican": 41, "asian": 60, "cafe": 26, "mealbox": 5,
    "chicken": 123, "mideast": 24, "indian": 30
}

# 이미지 URL 생성 함수
def get_image_url(category):
    directory = category_to_dir.get(category, "korean")
    max_index = dir_max_image_count.get(directory, 1)
    image_index = random.randint(1, max_index)
    return f"https://s3.studio1122.net/itseats/dummy_images/{directory}/{directory}_{image_index}.jpg"
```

```python
import requests
import json

def post_store(row):
    token = signup_and_login(row)
    if (token is None):
        print(f"[ERROR] {row['username']} 회원가입 또는 로그인 실패")
        return None

    url = "http://localhost:8080/api/owner/store-regist"

    # 랜덤한 배달비 설정
    random_fee = random.choice(range(1000, 3001, 500))

    store_request_data = {
        "name": row["사업장명"],
        "description": row["사업장명"],
        "address": row["도로명전체주소"],
        "lng": row["longitude"],
        "lat": row["latitude"],
        "phone": row["소재지전화"],
        "isFranchise": False,
        "categoryName": row["category"],
        "defaultDeliveryFee": random_fee,
        "onlyOneDeliveryFee": random_fee + 2000
    }

    files = {
        'request': ('request.json', json.dumps(store_request_data), 'application/json'),
        'images': ('store_image.txt', get_image_url(row["category"]), 'text/plain')  # 이미지 대신 문자열 파일
    }

    try:
        response = requests.post(url, files=files, headers={
            'Authorization': f'Bearer {token}',
        })
        return response.status_code
    except Exception as e:
        print(f"[ERROR] 매장 '{row['사업장명']}' 등록 실패: {e}")
        return None
    
# n번 and m번까지의 행에 대해 post_store 함수를 적용
n = 0
m = 1
df["status_code"] = df.loc[n:m].apply(post_store, axis=1)
```
