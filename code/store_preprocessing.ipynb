```python
import pandas as pd

file_path = 'fulldata_store.csv'
df = pd.read_csv(file_path, encoding='cp949')

print(df.head())
```

```python
print(df.columns)
```

```python
columns_to_keep = [
    '영업상태명',
    '소재지전화',
    '소재지전체주소',
    '도로명전체주소',
    '사업장명',
    '업태구분명',
    '좌표정보x(epsg5174)',
    '좌표정보y(epsg5174)',
    '위생업태명'
]

df = df[columns_to_keep]

print(df.head())
```

```python
status_counts = df['영업상태명'].value_counts()
print(status_counts)
```

```python
df = df[df['영업상태명'] == '영업/정상']
print(df.shape)  # (행 수, 열 수)
```

```python
status_counts = df['업태구분명'].value_counts()
print(status_counts)
```

```python
from pyproj import Transformer
import pandas as pd

# 좌표 변환기 설정: EPSG 5174 → EPSG 4326 (WGS84)
transformer = Transformer.from_crs("EPSG:5174", "EPSG:4326", always_xy=True)

# 결측치 제거: 사업장명과 좌표 필드
df = df.dropna(subset=['사업장명', '좌표정보x(epsg5174)', '좌표정보y(epsg5174)']).copy()

# 좌표 변환 함수
def convert_to_wgs84(row):
    x = row['좌표정보x(epsg5174)']
    y = row['좌표정보y(epsg5174)']
    lon, lat = transformer.transform(x, y)
    return pd.Series({'longitude': lon, 'latitude': lat})

# 위도/경도 컬럼 추가
df[['longitude', 'latitude']] = df.apply(convert_to_wgs84, axis=1)

# 결과 확인
print(df[['사업장명', '좌표정보x(epsg5174)', '좌표정보y(epsg5174)', 'longitude', 'latitude']].head())
```

```python
df.head()
```

```python
# CSV로 저장 (인코딩은 필요에 따라 cp949 또는 utf-8-sig)
df.to_csv('converted_store_data.csv', index=False, encoding='utf-8-sig')
```

```python
import pandas as pd

df = pd.read_csv('converted_store_data.csv', encoding='utf-8-sig')

print(df.head())
```

```python
# 결측치 제거: 사업장명과 좌표 필드
df2 = df.dropna(subset=['소재지전화']).copy()
```

```python
df2.head()
```

```python
# 첫 단어(시/도) 추출
df2['시도'] = df2['소재지전체주소'].str.split().str[0]

# 시/도별 개수 세기
region_counts = df2['시도'].value_counts()

print(region_counts)
```

```python
valid_regions = [
    '서울특별시', '부산광역시', '대구광역시', '인천광역시', '광주광역시',
    '대전광역시', '울산광역시', '세종특별자치시', '경기도', '강원특별자치도',
    '충청북도', '충청남도', '전북특별자치도', '전라남도',
    '경상북도', '경상남도', '제주특별자치도'
]

df2 = df2[df2['시도'].isin(valid_regions)].copy()
```

```python
print(df2['시도'].value_counts())
```

```python
df2.head()
```

```python
print(df2['업태구분명'].value_counts())
```

```python
exclude_types = ['기타 휴게음식점', '출장조리', '라이브카페', '키즈카페', '이동조리']
df2 = df2[~df2['업태구분명'].isin(exclude_types)].copy()

print(df2['업태구분명'].value_counts())
```

```python
# CSV로 저장 (인코딩은 필요에 따라 cp949 또는 utf-8-sig)
df2.to_csv('filtered_store_data.csv', index=False, encoding='utf-8-sig')
```
